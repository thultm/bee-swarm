```python
from pydub import AudioSegment
import math
import os
```

```python
class SplitWavAudioMubin():
    def __init__(self, folder, filename, overlap_sec=1, seconds_per_split=2):
        self.folder = folder
        self.filename = filename
        self.filepath = folder + '/' + filename
        self.overlap_sec = overlap_sec
        self.seconds_per_split = seconds_per_split
        self.audio = AudioSegment.from_wav(self.filepath)

    def get_duration(self):
        # Returns duration of audio in milliseconds
        return math.ceil(self.audio.duration_seconds) * 1000

    def audio_split(self):
        interval = self.seconds_per_split * 1000
        overlap = self.overlap_sec * 1000
        start = 0
        end = 0
        counter = 1
        
        # Check if duration is less than interval
        if self.get_duration() < interval:
            print('Duration of audio is less than specified interval')
            return
        
        # Iterate through audio and split at intervals
        for i in range(0, 2 * self.get_duration(), interval):
            if i == 0:
                start = 0
                end = interval
            else:
                start = end - overlap
                end = start + interval
            
            # If length of audio is less than interval, break
            if end >= self.get_duration():
                break

            # Filename to save
            split_audio_name = 'swarm' + '_' + self.filename.split('.')[0] + '_' + str(counter) + '.wav'
            split_audio_path = self.folder + '/' + split_audio_name

            # Split audio and save
            split_audio = self.audio[start:end]
            if split_audio.duration_seconds < 2:
                continue
            else:
                split_audio.export(split_audio_path, format="wav")
                print('Exporting: ' + split_audio_path)
                counter += 1
```

```python
folder = r'bee data\upload data\swarm'
file = [file for file in os.listdir(folder) if file.endswith('.WAV') or file.endswith('.wav')]
for f in file:
    split_wav = SplitWavAudioMubin(folder, f)
    print(split_wav.get_duration())
    split_wav.audio_split()
```

```python
# Get number of files (random) from a folder and copy to another folder
def remove_and_rename(folder, label):
    file = [file for file in os.listdir(folder) if file.endswith('.WAV') or file.endswith('.wav')]
    for f in file:
        if math.ceil(AudioSegment.from_wav(folder + '/' + f).duration_seconds) < 2:
            os.remove(folder + '/' + f)
            print('Removed: ' + f)
        else:
            # Rename label + index of file + .wav
            os.rename(folder + '/' + f, folder + '/' + label + '_' + str(file.index(f) + 1) + '.wav')
            print('Renamed: ' + f + ' to ' + label + '_' + str(file.index(f)) + '.wav')
```
